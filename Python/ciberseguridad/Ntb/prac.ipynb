### Renombrar

```python
df.columns={"Tipo_Ataque":"Ataque", "Tasa_Ocurrencia (%)":"Tasa"}
print(df.columns)
```

### Eliminar columna

```python
df.drop(["Vector_Infeccion"], axis=1, inplace=True)
```

### Pasar a dato entero

```python
df["Vector_Infeccion"]=df["Vector_Infeccion"].astype('int64')
```

### Si un dato es mayor a otro

```python

print(df[df["Nivel_Impacto (1-10)"]>8])
print(df[df["Costo_Recuperacion (USD)"]>5000])
print(df[df["Tasa_Ocurrencia (%)"]>20])
cinco_primeros=df[df["Tiempo_Recuperacion (horas)"]>100]
print(cinco_primeros.head(5))
```
