### Buscar valores máximos y mínimos:


```python
valor_minimo=df["Costo_Recuperacion (USD)"].min()
valor_maximo=df["Costo_Recuperacion (USD)"].max()
print(f'Valor mínimo: {valor_minimo}')
print(f'Valor máximo: {valor_maximo}')
```

### Valor con mayor relacion con el costo de recuperacion:

```python
print(df.corr(numeric_only=True))
# cerca de 1 indica una fuerte correlación positiva, cerca de -1 indica una fuerte correlación negativa, y cerca de 0 indica poca o ninguna correlación.
```
