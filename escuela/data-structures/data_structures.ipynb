En python hay 4 tipos básicos de estructuras de datos:
- Listas
- Tuplas
- Sets
- Dicccionarios

### Listas

Las listas son similares a los arreglos pero pueden contener elementos de distinto tipo.

```python
lista_nombres = ["Juan", "Carlos", "Laura", "Pedro"]
lista_nombres
```

```python
type(lista_nombres)
```

```python
print(lista_nombres[0])
print(lista_nombres[-1])
print(lista_nombres[:])
```

```python
#Insertar un elemento al final
lista_nombres.append("Jose")
lista_nombres
```

```python
#Insertar un elemento en una determinada posición
lista_nombres.insert(2, "Marta")
lista_nombres
```

```python
#Anexar un conjunto de elementos al final
lista_nombres.extend(["Pedro", "Juana", "Marcos"])
lista_nombres
```

```python
#Se pueden concatenar 2 listas con el operador sobrecargado "+"
lista_nombres2 = ["Brian", "Maria"]
lista_nombres3 = lista_nombres + lista_nombres2
lista_nombres3
```

```python
#Otro operador sobrecargado para listas es el "*" para repetición
lista_nombres4 = lista_nombres2 * 3
lista_nombres4
```

```python
#Obtener el índice de un determinado elemento (si está repetido devuelve el de su 1ra aparición)
lista_nombres.index("Pedro")
```

```python
#Número de repeticiones de un elemento
lista_nombres.count("Pedro")
```

```python
#Si quiero averiguar si un elemento está en la lista
if "Jose" in lista_nombres:
    print("yes")
```

```python
#Limpiar una lista
lista_nombres.clear()
len(lista_nombres)
```

```python
lista_nombres
```

```python
#Se dijo que las lista pueden contener diferentes tipos de datos
lista_nombres = ["Juan", True, 6, "Pedro"]
lista_nombres
```

```python
#operación simple usando indexación
lista_nombres[2] + lista_nombres[1]
```

### Eliminación de elementos

```python
#Es lo inverso a append, eliminando el último elemento
lista_nombres.pop()
lista_nombres
```

```python
#Para eliminar un elemento específico
lista_nombres.remove(True)
lista_nombres
```

### Listas anidadas

```python
movies = ["The Holy Grail", 1975, "Terry Jones & Terry Gilliam", 91,
   ...: ["Graham Chapman", ["Michael Palin", "John Cleese",
   ...: "Terry Gilliam", "Eric Idle", "Terry Jones"]]]
movies
```

```python
type(movies[1])
```

```python
#Si quiero displayar los elementos basta con hacer
for elem in movies:
    print(elem)
```

```python
#Pero algunos de sus elementos son listas completas...deberiamos poder mostrar tambien sus elementos
for elem in movies:
    if isinstance(elem, list):
        for innerelem in elem:
            if isinstance(innerelem, list):
                for innerinnerelem in innerelem:
                    print(innerinnerelem)
            else:
                print(innerelem)
    else:
        print(elem)
```

### Tuplas

Una tupla es muy similar a una lista pero es inmutable. Es decir que una vez definida no puede cambiar de tamaño (no append, no pop) y ni siquiera modificar su contenido. Se define con sus elemento entre "( )"

```python
mi_tupla = ("Jose", 34, 1.75, True)
print(type(mi_tupla))
mi_tupla
```

```python
mi_tupla[-1]
```

```python
#Se puede convertir una tupla en una lista y viceversa utilizando: list y tuple
nueva_lista = list(mi_tupla)
print(type(nueva_lista))
nueva_tupla = tuple(nueva_lista)
print(type(nueva_tupla))
nueva_lista
```

```python
#Las funciones que no alteran el tamaño de las listas son aplicable también para tuplas
print(mi_tupla.count("Jose"))
print(mi_tupla.index(1.75))
```

```python
#Puedo conocer la extensión de una lista o tupla utilizando len
print(len(nueva_lista))
print(len(nueva_tupla))
```

### Sets o Conjuntos

Están entre las listas y las tuplas. Permiten agregar o quitar elementos pero no modificarlos. No soportan indexación.

```python
# Conjunto de colores únicos
conjunto_colores = {'rojo', 'verde', 'azul'}
```

```python
conjunto_colores.add('negro')
```

```python
conjunto_colores.add(5)
```

```python
conjunto_colores.discard('verde')
```

```python
conjunto_colores.discard('rojo')
```

```python
conjunto_colores
```

```python
conjunto_colores.remove('rojo')
```

```python
conjunto_colores
```

```python
conjunto_colores.clear()
```

```python
conjunto_colores
```

```python
conjunto_colores = {'rojo', 'verde', 'azul'}
```

```python
conjunto_par = {'rojo', 'verde', 3}
```

```python
conjunto_colores.union(conjunto_par)
```

### Diccionarios

Son estructuras de datos parecidas a las listas y las tuplas pero como peculiaridad tienen que cada elemento es una asociación clave-valor, donde la clave es única.

```python
mi_dict = {"Francia":"Paris", "Argentina":"Buenos Aires", "Brasil":"Brasilia", "Peru":"Lima"}
mi_dict
```

```python
#Para obtener un elemento basta con hacer referencia a su clave
mi_dict["Brasil"]
```

```python
#Agrego un elemento solo haciendo referencia a una clave nueva
mi_dict["Ecuador"] = "Quito"
mi_dict
```

```python
#Modifico un elemento sobreescribiendo su valor y usando la misma clave
mi_dict["Brasil"] = "Rio de Janiero"
mi_dict
```

```python
#Elimino un elemento del diccionario
del mi_dict["Peru"]
mi_dict
```

```python
#Puedo listar claves y valores por separados
print(mi_dict.keys())
print(mi_dict.values())
print(len(mi_dict))
```

```python
#A su vez el diccionario puede contener diferentes tipos de datos
mi_dict = {"Nombre": "Jose", "Edad":34, "Vive": True}
mi_dict
```

```python
#Se pueden tener diccionarios anidados
mi_dict = {"Nombre": "Jose", "Edad":34, "Vive": True, "CV":{"Nombre": "Jose", "edad": 36}}
```

```python
mi_dict
```

```python
type(mi_dict["CV"])
```

```python
mi_dict["CV"]["edad"]
```
